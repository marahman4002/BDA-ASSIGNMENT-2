# 1.Titanic Classification Model (Logistic Regression with PySpark)

## ✅ Why Titanic Dataset?
The Titanic dataset is a classic example for binary classification. It contains both numerical and categorical features like age, gender, fare, and class, which makes it great for a beginner-friendly ML workflow.

## ✅ Why Logistic Regression?
Logistic Regression is one of the simplest and most interpretable classification models. It is particularly suitable for binary outcomes like "Survived or Not Survived". It also performs well on linearly separable data and helps to understand feature influence.

## ✅ Key Steps
1. Cleaned and selected relevant features (Pclass, Sex, Age, etc.)
2. Converted categorical feature (`Sex`) to numerical.
3. Used `VectorAssembler` to combine features.
4. Trained logistic regression on 70% of the data.
5. Evaluated on 30% test set using accuracy metric.

## ✅ Accuracy
Our simple model achieves around 75–80% accuracy on the test set, which is decent for such a basic model without feature engineering or hyperparameter tuning.


```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
import pandas as pd

pdf = pd.read_csv(url)
pdf.head()
```

```python
df = spark.createDataFrame(pdf)
df.printSchema()
df.show(5)
```

```python
from pyspark.sql.functions import col
df = df.select("Survived", "Pclass", "Sex", "Age", "SibSp", "Parch", "Fare")
df = df.dropna()
```

```python
from pyspark.ml.feature import StringIndexer

indexer = StringIndexer(inputCol="Sex", outputCol="SexIndex")
df = indexer.fit(df).transform(df)
```

```python
from pyspark.ml.feature import VectorAssembler

assembler = VectorAssembler(
    inputCols=["Pclass", "SexIndex", "Age", "SibSp", "Parch", "Fare"],
    outputCol="features"
)

df = assembler.transform(df)
df = df.select("features", "Survived")
```

```python
from pyspark.ml.classification import LogisticRegression

train, test = df.randomSplit([0.7, 0.3], seed=42)
lr = LogisticRegression(labelCol="Survived", featuresCol="features")
model = lr.fit(train)
```

```python
predictions = model.transform(test)
predictions.select("Survived", "prediction", "probability").show(10)
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

evaluator = MulticlassClassificationEvaluator(
    labelCol="Survived", predictionCol="prediction", metricName="accuracy"
)

accuracy = evaluator.evaluate(predictions)
print(f"Test Accuracy = {accuracy:.2f}")
```

# 2.Iris Clustering Model (KMeans with PySpark)

## ✅ Why Iris Dataset?
The Iris dataset is a classical dataset for unsupervised learning and clustering. It contains four numerical features (length and width of sepals and petals) which make it ideal for KMeans clustering. Even though species labels are available, we don’t use them during clustering.

## ✅ Why KMeans?
KMeans is a fast and efficient clustering algorithm that partitions the dataset into **k distinct clusters** based on feature similarity. Since Iris has 3 species, it makes sense to try `k=3`.

## ✅ Key Steps
1. Loaded the Iris dataset and selected numerical features.
2. Used `VectorAssembler` to prepare data for Spark ML.
3. Trained KMeans model with `k=3`.
4. Evaluated with WSSSE and printed cluster centers.

## ✅ Cluster Quality

- We used the **Silhouette Score** to evaluate cluster quality (range: -1 to 1).
- A **higher silhouette score** indicates better-defined and well-separated clusters.
- This metric considers both **intra-cluster cohesion** and **inter-cluster separation**.



```python
import pandas as pd

url = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
pdf = pd.read_csv(url)
pdf.head()
```

```python
df = spark.createDataFrame(pdf)
df.printSchema()
df.show(5)
```

```python
from pyspark.ml.feature import VectorAssembler

assembler = VectorAssembler(
    inputCols=["sepal_length", "sepal_width", "petal_length", "petal_width"],
    outputCol="features"
)

iris_df = assembler.transform(df)
iris_df = iris_df.select("features")  # drop label for unsupervised learning
iris_df.show(5, truncate=False)
```

```python
from pyspark.ml.clustering import KMeans

kmeans = KMeans(k=3, seed=1)
model = kmeans.fit(iris_df)

predictions = model.transform(iris_df)
predictions.show(10)
```

```python
from pyspark.ml.evaluation import ClusteringEvaluator

# Get predictions from the KMeans model
predictions = model.transform(iris_df)

# Create evaluator for clustering using the "silhouette" metric
evaluator = ClusteringEvaluator(predictionCol='prediction', featuresCol='features', metricName='silhouette', distanceMeasure='squaredEuclidean')

# Calculate silhouette score
silhouette = evaluator.evaluate(predictions)
print(f"Silhouette Score = {silhouette:.4f}")
```

```python
centers = model.clusterCenters()
print("Cluster Centers:")
for c in centers:
    print(c)
```

# 3.MovieLens Recommender (ALS with PySpark)

## ✅ Why MovieLens Dataset?
The MovieLens dataset is the most popular dataset for building recommendation systems. It contains real-world movie ratings from thousands of users.

## ✅ Why ALS (Alternating Least Squares)?
ALS is Spark's go-to algorithm for collaborative filtering. It decomposes the user-item interaction matrix into two lower-rank matrices and uses matrix factorization to predict missing entries (i.e., unrated movies).

## ✅ Key Steps
1. Loaded and cleaned the MovieLens 100K ratings data.
2. Used Spark's ALS model with user, item, and rating columns.
3. Trained the model on 80% of data and tested on 20%.
4. Evaluated model using RMSE.
5. Generated top-N movie recommendations per user.

## ✅ Output
- RMSE shows how accurate the rating predictions are.
- Recommendations include top 5 movies per user that they haven’t rated yet, based on model predictions.

## ✅ Notes
- ALS works well for large-scale, sparse matrices (like user-item ratings).
- Cold start strategy is set to "drop" to avoid NaNs during testing.


```python
url = "https://files.grouplens.org/datasets/movielens/ml-100k/u.data"
import pandas as pd

columns = ["userId", "movieId", "rating", "timestamp"]
ratings_pd = pd.read_csv(url, sep="\t", names=columns)
ratings_pd.drop("timestamp", axis=1, inplace=True)
ratings_pd.head()
```

```python
ratings = spark.createDataFrame(ratings_pd)
ratings.printSchema()
ratings.show(5)
```

```python
from pyspark.ml.recommendation import ALS

(training, test) = ratings.randomSplit([0.8, 0.2], seed=42)

als = ALS(
    userCol="userId",
    itemCol="movieId",
    ratingCol="rating",
    nonnegative=True,
    coldStartStrategy="drop"
)

model = als.fit(training)
```

```python
predictions = model.transform(test)
predictions.show(10)
```

```python
from pyspark.ml.evaluation import RegressionEvaluator

evaluator = RegressionEvaluator(
    metricName="rmse",
    labelCol="rating",
    predictionCol="prediction"
)

rmse = evaluator.evaluate(predictions)
print(f"Root-mean-square error = {rmse:.2f}")
```

```python
user_recs = model.recommendForAllUsers(5)
user_recs.show(3, truncate=False)
```

```python
user_id = 42
model.recommendForUserSubset(ratings.filter(f"userId = {user_id}").select("userId").distinct(), 5).show(truncate=False)
```
